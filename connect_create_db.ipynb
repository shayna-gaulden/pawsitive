Function to insert data

```python
def insertdata(data,table):
    s = ["%s,"]*data.shape[1]
    s = "".join(s)
    s = s[0:-1]
    
    for i,row in data.iterrows():
        sql = "INSERT INTO " + table + " VALUES (" + s + ")"
        print(sql)
        cursor.execute(sql, tuple(row))
        print("Record inserted")
        cnxn.commit()
```
